```python
import subprocess


# Run the TransCoder command and capture the output
cmd = "python -m codegen_sources.model.translate --src_lang cpp --tgt_lang java --model_path 'TransCoder_model_1.pth' --beam_size 1 < hello.cpp"
process = subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
stdout, stderr = process.communicate()

# Function to extract and print the C++ and Java code
def extract_and_print_code(output):
    # Markers to identify the start of C++ and Java code in the output
    cpp_start_marker = "Input cpp function:"
    java_start_marker = "Translated java function:"

    # Find the start of C++ and Java code
    cpp_start_index = output.find(cpp_start_marker)
    java_start_index = output.find(java_start_marker)

    # Extract the C++ and Java code
    cpp_code = output[cpp_start_index:java_start_index].strip()
    java_code = output[java_start_index:].strip()

    # Clean up and isolate the code
    cpp_code_clean = cpp_code.replace(cpp_start_marker, '').strip()
    java_code_clean = java_code.replace(java_start_marker, '').strip().replace("====================", '').strip()

    # Print the C++ and Java code
    print("C++ Code:\n", cpp_code_clean)
    print("\nJava Code:\n", java_code_clean)

    return cpp_code_clean, java_code_clean

# Call the function to extract and print the code
cpp_code, java_code = extract_and_print_code(stdout)
```
